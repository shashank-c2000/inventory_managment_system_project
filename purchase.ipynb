```python
##Reading inventory
import json
fd = open("my_record.json",'r')
r = fd.read()
fd.close()

record= json.loads(r)
record

```

```python
##Reading sales
import json
fd = open("sales.json",'r')
t = fd.read()
fd.close()

record1= json.loads(t)
record1
```

```python
flag=0
while(flag==0):
  ui_prod  = str(input("Enter the product_Id: "))
  ui_quant = int(input("Enter the quantity: "))

  if(ui_quant<= record[ui_prod]['qt']):
    print("Product: ", record[ui_prod]['name'])
    print("Price: ", record[ui_prod]['pr'])
    print("Billing Amount: ", record[ui_prod]['pr'] * ui_quant)

    record[ui_prod]['qt'] = record[ui_prod]['qt'] - ui_quant

#update inentory after purchase

    js = json.dumps(record)

    fd = open("my_record.json",'w')
    fd.write(js)
    fd.close()
  #update to sales file

    record1[ui_prod] = {'name': record[ui_prod]['name'], 'pr': record[ui_prod]['pr'], 'qt': ui_quant,'weight':record[ui_prod]['weight'],'manufacture':record[ui_prod]['manufacture'],'Billing amount':record[ui_prod]['pr'] * ui_quant}
    js = json.dumps(record1)

    fd = open("sales.json",'w')
    fd.write(js)
    fd.close()
    flag=1
  else:
    print("Availaible max quantity is",record[ui_prod]['qt'])
    print("Please enter valid input")
```
