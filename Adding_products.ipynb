```python
##Reading inventory
import json
fd = open("my_record.json",'r')
r = fd.read()
fd.close()

record= json.loads(r)
record
```

```python

#Adding new product to inventory
prod_id = str(input("Enter product id:"))
name = str(input("Enter name:"))
pr = int(input("Enter price:"))
qn = int(input("Enter quantity:"))
wg = str(input("Enter weight"))
md = str(input("Enter the manufacture date"))


record[prod_id] = {'name': name, 'pr': pr, 'qt': qn,'weight':wg,'manufacture':md}

js = json.dumps(record)

fd = open("my_record.json",'w')
fd.write(js)
fd.close()
```
